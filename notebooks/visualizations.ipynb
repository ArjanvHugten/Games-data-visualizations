```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

%matplotlib inline

game_sales = pd.read_csv(r'vgsales.csv', sep=',', low_memory=False, encoding = 'ISO-8859-1', header=0)

game_sales.head()
```

Above you can see the first 5 rows and the last 5 rows of the game sales dataset. The dataset has a rank, which is the rank of game sales. The name of the game. The platform the game is released for. Genre of the game. Publisher. And the sales in NA, EU, JP, Other and the global sales.

First I will visualize what platforms the games are on. I will do this with a bar chart. 

```python
platform_count  = game_sales['Platform'].value_counts()

#print(platform_count)

plt.figure(figsize=(10,5))
plt.bar(platform_count.index, platform_count.values)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Platform', fontsize=12)
plt.xticks(rotation=90)    
plt.show()
```

As you can see DS and PS2 have the most games. A good question is if the games for the consoles have a better sale with the amount of games.

```python
platform_sales  = game_sales.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

#print(platform_sales)

plt.figure(figsize=(10,5))
plt.bar(platform_sales.index, platform_sales.values)
plt.ylabel('Total copies sold in millions', fontsize=12)
plt.xlabel('Platform', fontsize=12)
plt.xticks(rotation=90)    
plt.show()
```

As you can see the amount of games doesn't need to be important in the total sales for a console. The DS has the most games and is just the 5th in total sales. The PS2 had the second to most games and also has very good sales. A new question that raised after this graph is what genre is the most popular under the games. 

```python
genre_sales  = game_sales.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

#print(genre_sales)

plt.figure(figsize=(10,5))
plt.bar(genre_sales.index, genre_sales.values)
plt.ylabel('Total copies sold in millions', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=90)    
plt.show()
```

The genre with the most sale is of course action. An interesting runner up which I didn't expect is sports. The third is shooters which I understand. Another interesting graph to see is the year in which games are released. With that graph we can see what year has the most games and what the trend is in the release of games.

```python
data = game_sales[game_sales['Year'] <= 2016]
                  
games_year = data.groupby('Year', sort=True).size()

plt.figure(figsize=(10,5))
plt.plot(games_year.index, games_year.values)
plt.show()
```

As you can see the number of games grows during the whole period 1980 untill somewhere around 2009. Then a massive drop starts. The PS2 is released in 2000 and the DS in 2004. During these periods you can see the number of games grows this adds up that the DS and PS2 have the most games.

An interesting question is what the most popular genre is per regio. To do this I wanted to create a parallel coordinates graph.

```python
from pandas.plotting import parallel_coordinates

cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
data = game_sales[game_sales['Rank'] <= 1000].groupby('Genre', as_index=False)[cols].sum()

plt.figure(figsize=(10,5))
parallel_coordinates(data, class_column='Genre', cols=cols)
```

Above graph shows the sales of different genre's in different regions. The sales of Action are very heigh in NA and a lot less heigh in japan the same goes for Shooter.

```python
region_sales = pd.DataFrame(columns=['Region','Sales'])
region_sales.loc[1] = ['NA', game_sales['NA_Sales'].sum()]
region_sales.loc[2] = ['JP', game_sales['JP_Sales'].sum()]
region_sales.loc[3] = ['EU', game_sales['EU_Sales'].sum()]
region_sales.loc[4] = ['Other', game_sales['Other_Sales'].sum()]

plt.figure(figsize=(10,5))
plt.bar(region_sales['Region'], region_sales['Sales'])
plt.xticks(rotation=90)    
plt.show()
```
